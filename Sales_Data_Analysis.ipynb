```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate realistic sample sales data
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=500, freq='D')

data = {
    'Order_ID': range(1001, 1501),
    'Order_Date': np.random.choice(dates, 500),
    'Category': np.random.choice(['Electronics', 'Fashion', 'Home & Kitchen', 'Beauty', 'Sports'], 500),
    'Product': np.random.choice(['Laptop', 'T-Shirt', 'Blender', 'Cream', 'Football'], 500),
    'Quantity': np.random.randint(1, 10, 500),
    'Unit_Price': np.random.uniform(10, 500, 500).round(2),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 500),
    'Customer_Type': np.random.choice(['New', 'Returning'], 500)
}

df = pd.DataFrame(data)
df['Sales'] = (df['Quantity'] * df['Unit_Price']).round(2)

print("✅ Sample Sales Data Created Successfully!")
print("Shape of dataset:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

# Save as CSV file
df.to_csv('sales_data.csv', index=False)
print("\n📁 File saved as 'sales_data.csv'")

# Check basic information
print(df.info())
print("\nStatistical Summary:")
display(df.describe())

# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Sales by Category
print(df['Category'].value_counts())

# Data Overview
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())

# Summary Statistics
display(df.describe())

# Total Sales
total_sales = df['Sales'].sum()
print(f"\nTotal Sales: ₹{total_sales:,.2f}")

# Average Order Value
avg_order = df['Sales'].mean()
print(f"Average Order Value: ₹{avg_order:.2f}")

# Sales by Category
category_sales = df.groupby('Category')['Sales'].agg(['sum', 'count']).round(2)
print("Sales by Category:")
display(category_sales)

# Sales by Region
region_sales = df.groupby('Region')['Sales'].sum().round(2)
print("\nSales by Region:")
display(region_sales)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Total Sales by Category', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Total Sales (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_by_category.png')
plt.show()

# Monthly Sales Trend
df['Month'] = df['Order_Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
monthly_sales['Month'] = monthly_sales['Month'].astype(str)

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], marker='o', linewidth=2.5)
plt.title('Monthly Sales Trend (2023)', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Total Sales (₹)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('monthly_sales_trend.png')
plt.show()

plt.figure(figsize=(8, 8))
region_sales = df.groupby('Region')['Sales'].sum()
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Sales Distribution by Region')
plt.savefig('sales_by_region.png')
plt.show()

plt.figure(figsize=(8, 8))
region_sales = df.groupby('Region')['Sales'].sum()
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Sales Distribution by Region')
plt.savefig('sales_by_region.png')
plt.show()

plt.figure(figsize=(10, 6))
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(5)
sns.barplot(x=top_products.values, y=top_products.index, palette='mako')
plt.title('Top 5 Products by Sales')
plt.xlabel('Total Sales (₹)')
plt.tight_layout()
plt.savefig('top_products.png')
plt.show()


# Customer Type Analysis
plt.figure(figsize=(8, 6))
customer_sales = df.groupby('Customer_Type')['Sales'].agg(['sum', 'count']).round(2)
print("Customer Type Performance:")
display(customer_sales)

sns.barplot(x=customer_sales.index, y=customer_sales['sum'], palette='coolwarm')
plt.title('Total Sales by Customer Type')
plt.ylabel('Total Sales (₹)')
plt.savefig('customer_type.png')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Quantity', y='Sales', hue='Category', alpha=0.7)
plt.title('Quantity vs Sales Amount')
plt.savefig('quantity_vs_sales.png')
plt.show()

print("🔍 KEY BUSINESS INSIGHTS\n")

total_sales = df['Sales'].sum()
print(f"1. Total Revenue Generated: ₹{total_sales:,.2f}")

top_category = df.groupby('Category')['Sales'].sum().idxmax()
print(f"2. Top Performing Category: {top_category}")

avg_order = df['Sales'].mean()
print(f"3. Average Order Value: ₹{avg_order:.2f}")

returning_sales = df[df['Customer_Type'] == 'Returning']['Sales'].sum()
print(f"4. Returning Customers Contribution: ₹{returning_sales:,.2f} ({returning_sales/total_sales*100:.1f}%)")

print("\n5. Recommendations:")
print("- Focus more marketing on top category:", top_category)
print("- Target New Customers with better offers to increase their contribution")
print("- Potential to boost sales by 12-18% by improving product mix and customer retention")
```
